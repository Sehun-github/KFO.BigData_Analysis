## 막대그래프 실습
* 막대 폭 지정
* 막대 색상 지정
* 막대 테두리
* 막대 패턴
* 막대 위치
* 두개의 막대 그래프 비교

```python
import pandas as pd
import matplotlib.pyplot as plt

# 그래프에 한글 설정
plt.rc('font',family='Malgun Gothic')

# 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터 df1, df2
df1 = pd.DataFrame({'요일':['월','화','수','목','금','토','일'], 
                   '매출액':[10000,9000,11000,8000,13000,15000,14000]})
df2 = pd.DataFrame({'요일':['월','화','수','목','금','토','일'], 
                   '매출액':[9000,9500,13000,7000,12000,14000,11000]})
```

```python
plt.bar(df1['요일'],df1['매출액'])
plt.show()
```

### 막대 그래프 폭 조정
* 가로막대 그래프 width = 0~1(default:0.8)
* 세로막대 그래프 height=0~1(default:0.8)

```python
plt.bar(df1['요일'],df1['매출액'], width = 0.5)
plt.show()
```

```python
plt.barh(df1['요일'],df1['매출액'], height = 0.5)
plt.show()
```

### 막대 색상 지정
* color='색깔'
* 막대마다 색깔: color=['색상','색상']

```python
plt.bar(df1['요일'],df1['매출액'], width = 0.5, color=['r','orange','y','g','b','navy','violet'])
plt.show()
```

### 막대 테두리
* edgecolor = 테두리 색상
* linewidth = 테두리 두께

```python
plt.bar(df1['요일'],df1['매출액'], width = 0.5, color=['r','orange','y','g','b','navy','violet'],
                                   edgecolor='gray',linewidth=2)
plt.show()
```

### 막대 패턴
* hatch 파라미터에 기호 전달 : '/', '\', '|', '-', '+', 'x', 'o', 'O', '.', '*'
* 패턴기호의 개수로 밀도를 조정한다.
* 막대 마다 패턴 다르게 할때: set_hatch(기호)

```python
plt.bar(df1['요일'],df1['매출액'], width = 0.5, color=['r','orange','y','g','b','navy','violet'],
                                   edgecolor='gray',linewidth=2, hatch='xx')
plt.show()
```

```python
bars = plt.bar(df1['요일'],df1['매출액'], width = 0.5, color=['r','orange','y','g','b','navy','violet'],
                                   edgecolor='gray',linewidth=2)

bars[0].set_hatch('.')
bars[1].set_hatch('/')
bars[2].set_hatch('+')
bars[3].set_hatch('-')
bars[4].set_hatch('*')
bars[5].set_hatch('x')
plt.show()
```

### 막대 위치
* align = center/edge
* edge로 지정하면 막대의 왼쪽 끝과 틱을 맞춘다
* 막대의 오른쪽 끝과 틱을 맞추려면 width를 음수로 지정한다.

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4, align='edge')
plt.show()
```

```python
# width를 음수로 지정하여 오른쪽 끝과 틱을 맞춘다
plt.bar(df1['요일'],df1['매출액'], width=-0.4, align='edge')
plt.show()
```

### 두개의 막대 그래프 비교

```python
# width - defalut가 0.8이기 때문에 0.4/0.4로 나누어 표현한다.
plt.bar(df1['요일'],df1['매출액'], width=-0.4, align='edge')
plt.bar(df2['요일'],df2['매출액'], width=0.4,align='edge')
plt.title('두 매장의 요일별 매출액 평균', size = 15)
plt.grid(axis='y',ls=':')
plt.show()
```
