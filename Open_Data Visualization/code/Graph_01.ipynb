## 기본 그래프 그리기 (matplotlib 사용)
1. 리스트
2. 시리즈
3. 데이터프레임

```python
import matplotlib.pyplot as plt
```

### 리스트로 그리기

```python
# 리스트로 그리기
data = [1,5,3,9,7]
plt.plot(data)
plt.show()
```

### 시리즈로 그리기

```python
# 시리즈로 그리기
import pandas as pd
s = pd.Series([1,5,3,9,7], index=['a','b','c','d','f']) # index설정(a~f), 인덱스와 value크기가 같아야함
plt.plot(s)
plt.show()
```

```python
# 리스트 - x축 y축을410 
x = [1,5,6,9,10]
y = [1,5,3,9,7]

# 그래프 그리기
plt.plot(x,y)
plt.show()
```

### 데이터프레임으로 그리기

```python
# 데이터 프레임으로 그리기
df = pd.DataFrame({'x':[1,5,6,9,10], 'y':[1,5,3,9,7]})
plt.plot(df['x'],df['y'])
plt.show()
```
