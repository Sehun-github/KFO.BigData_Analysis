```python
import matplotlib.pyplot as plt
```

### 선 그래프
* plt.plot(x,y)

```python
age = [7,8,9,10,11,12,13,14,15]
height = [122.1, 127.9, 133.4, 138.8, 142.7, 151.4, 155.6, 165.0, 169.2]
plt.plot(age,height)
plt.show()
```

### 막대 그래프
* plt.bar(x,y)

```python
blood_type=['A','B','O','AB']
count=[45,35,15,5]
plt.bar(blood_type,count)
plt.show()
```

#### 가로막대 그래프
* plt.barh(x,y)

```python
plt.barh(blood_type,count) # bar -> barh
plt.show()
```

### 파이차트
* plt.pie(y, labels=x)

```python
blood_type=['A','B','O','AB']
count=[45,35,15,5]
plt.pie(count, labels=blood_type)
plt.show()
```

### 히스토그램
* plt.hist(y, bins = ???) //  bins - 구간수

```python
# 샘플데이터 (1~100사이의 랜덤 정수 1000개)
import numpy as np
data = np.random.randint(1,101,1000)
plt.hist(data, bins=100) # bins = 구간수
plt.show()
```

### 상자수염 그래프
* plt.boxplot(y)

```python
# 샘플데이터 (1~100사이의 랜덤 정수 1000개)
import numpy as np
data = np.random.randint(1,101,1000)
plt.boxplot(data) # 가운데 주황줄 - Median
plt.show()
```

### 바이올린 그래프
* plt.violinplot(y)

```python
# 샘플데이터 (1~100사이의 랜덤 정수 1000개)
import numpy as np
data = np.random.randint(1,101,1000)
plt.violinplot(data)
plt.show()
```

### 산점도
* plt.scatter(x,y)

```python
import seaborn as sns
tips = sns.load_dataset('tips')[['total_bill','tip']] # seaborn을 통해 데이터 불러오기
plt.scatter(tips['total_bill'],tips['tip'])
plt.show()
```
