## 히스토그램
* plt.hist(data)
* bins=구간의 개수
* 누적 히스토그램: cumulative=True
* 범위지정: range=(min,max)
* 밀도표시: density=True
* 가로 히스토그램: orientation='horizontal'
* 선만 표시 histtype='step'
* style 속성

````python
import pandas as pd
import matplotlib.pyplot as plt
# 그래프에 한글 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 유니코드에서 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False
````

````python
# 데이터
scores = [0,10,15,15,15,16,19,20,21,25,25,26,26,29,30,35,36,37,39,40,41,41,44,45,45,45,45,47,
          50,50,50,50,51,51,51,53,54,55,55,56,60,61,62,62,63,64,65,65,65,65,66,66,66,66,66,
          67,68,68,69,70,70,70,70,70,70,70,70,71,71,71,71,71,72,72,72,72,73,74,74,74,75,75,
          76,76,76,77,77,77,77,78,78,78,78,78,79,79,79,79,80,80,80,80,80,80,81,81,81,82,82,
          85,85,85,88,88,89,90,90,90,93,93,95,95,95,97,100]
````

### 구간, 범위 설정

````python
# 50~90까지 범위의 20개 구간을 나눈 히스토그램
plt.hist(scores, bins=20, range=(50,90))
plt.show()
````

### 누적 히스토그램

````python
plt.hist(scores, cumulative=True)
plt.show()
````

### 밀도표시
*  왼쪽 Y축이 밀도로 바뀜

````python
plt.hist(scores, density=True)
plt.show()
````

### 가로 히스토그램

````python
plt.hist(scores, orientation='horizontal')
plt.show()
````

### 선만 표시
* histtype='step'

````python
plt.hist(scores, histtype='step')
plt.show()
````

### style 속성
```
 - rwidth=막대폭(0~1)
 - color=막대색
 - alpha=투명도
 - edgecolor(ec)=선색 
 - linewidth(lw)=선두께
 - linestyle(ls)=선스타일
 - hatch=패턴
```

````python
plt.hist(scores, rwidth=0.9, color='b', alpha=0.5, ec='k', lw=2,ls=':',hatch='/')
plt.xticks(range(0,101,10))
plt.grid(axis='y',ls=':')
plt.xlabel('점수')
plt.ylabel('인원')
plt.title('점수대별 인원분포', size=20)
plt.show()
````
