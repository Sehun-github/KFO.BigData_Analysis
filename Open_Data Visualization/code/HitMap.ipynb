## 히트맵
* pyplot으로 히트맵 그리기
* seaborn으로 히트맵 그리기

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#데이터
arr = np.random.standard_normal((5,5))
df = pd.DataFrame(arr)
df
```

### pyplot으로 히트맵 그리기
* 컬러맵 지정 :   cmap='색상'
* 컬러바 지정 : plt.colorbar()

```python
plt.pcolor(df, cmap='Blues')
plt.colorbar()
plt.show()
```

### seaborn
* heatmap(data=2차원데이터)
* cmap=컬러맵 : 컬러맵 지정
* annot=True : 수치 표시
* fmt='d' : 정수로 표시

```python
import seaborn as sns
sns.heatmap(df, cmap='Blues', annot=True)
plt.show()
```
