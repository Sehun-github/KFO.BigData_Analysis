## 여러 그래프 그리기 & 범례 설정

````python
import matplotlib.pyplot as plt
import matplotlib as mpl
# 그래프에 한글 설정
mpl.rcParams['font.family'] = 'NanumSquare'

# 그래프에 마이너스 기호 깨지는 문제 해결
mpl.rcParams['axes.unicode_minus'] = False
````

````python
import pandas as pd
df1 = pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[80,78,75,73,70]})
df2 = pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[60,62,59,55,54]})
````

### 여러 그래프 그리기

````python
plt.plot(df1['월'],df1['몸무게'])
plt.plot(df2['월'],df2['몸무게'])
plt.show()
````

### 범례 설정
* plt.legend(loc=위치번호)
* plt.legend(loc=위치이름)  
```
        Location String   Location Code
        ===============   =============
        'best'            0
        'upper right'     1
        'upper left'      2
        'lower left'      3
        'lower right'     4
        'right'           5
        'center left'     6
        'center right'    7
        'lower center'    8
        'upper center'    9
        'center'          10
        ===============   =============
```
* plt.legend(loc=(x위치,y위치)) : 범례의 왼쪽 하단이 기준점.

````python
# X축 틱 라벨
def fn_xtick(x):
    x = str(x)
    if x=='0' or x=='13':
        return ''
    else:
        return x+'월'

xtick_label = pd.Series(range(0,14,1))
xtick_label = xtick_label.apply(fn_xtick)
````

````python
# y축 틱 라벨
def fn_ytick(x):
    x = str(x)
    return x+'kg'

ytick_label = pd.Series(range(30,95,5))
ytick_label = ytick_label.apply(fn_ytick)
````

````python
# 두 그래프 그리기
plt.plot(df1['월'],df1['몸무게'], label='James')
plt.plot(df2['월'],df2['몸무게'], label='Amy')
plt.legend(loc=(1.01,0.8))

# 축 라벨
plt.xlabel('월', fontsize=12)
plt.ylabel('몸무게(kg)', fontsize=12)

# x축 틱 : 1월, 2월, 3월,...,12월
plt.xticks(range(0,14,1), labels=xtick_label)

# y축 틱 : 30kg, 35kg, 40kg, ..., 90kg
plt.yticks(range(30,95,5), labels=ytick_label)

# 그리드
plt.grid(ls=':')

plt.show()
````

#### 범례 속성 지정
* 열 개수 : ncol=열개수
* 폰트 사이즈 : fontsize=폰트사이즈
* 테두리 : frameon=(True/False)
* 음영 : shadow=(True/False)
* 바탕색 : facecolor=색상
* 테두리색 :edgecolor=색상

````python
# 두 그래프 그리기
plt.plot(df1['월'],df1['몸무게'], label='James')
plt.plot(df2['월'],df2['몸무게'], label='Amy')

plt.legend(ncol=2,fontsize=12,shadow=True,facecolor='ivory',edgecolor='k') # 범례 설정

plt.title('월별 몸무게 변화', size=15)
plt.xlabel('월', fontsize=12)
plt.ylabel('몸무게(kg)', fontsize=12)

plt.xticks(range(0,14,1), labels=xtick_label)
plt.yticks(range(30,95,5), labels=ytick_label)

# 그리드
plt.grid(ls=':')

plt.show()
````
