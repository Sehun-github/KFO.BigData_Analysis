## 서브 플롯

```python
import matplotlib.pyplot as plt
```

```python
# 데이터 불러오기
import seaborn as sns
anscombe = sns.load_dataset('anscombe')
anscombe.head(3)
```

```python
df1 = anscombe[anscombe['dataset']=='I']
df2 = anscombe[anscombe['dataset']=='II']
df3 = anscombe[anscombe['dataset']=='III']
df4 = anscombe[anscombe['dataset']=='IV']
```

### 4개의 그래프를 한번에 그리기

```python
plt.plot(df1['x'],df1['y'],'o')
plt.plot(df2['x'],df2['y'],'o')
plt.plot(df3['x'],df3['y'],'o')
plt.plot(df4['x'],df4['y'],'o')
plt.show()
```

### 4개의 그래프를 각각 그리기
* plt.figure(figsize=(x사이즈, y사이즈) -- 전체 도화지 크기 설정
* plt.subplot(전체행개수,전체열개수,그래프순서)

```python
fig = plt.figure(figsize=(9,6), facecolor='ivory') # 전체 도화지 설정

plt.subplot(221)
plt.plot(df1['x'],df1['y'],'o')
plt.title('ax1') # 서브 타이틀 제목

plt.subplot(222)
plt.plot(df2['x'],df2['y'],'+')
plt.title('ax2') # 서브 타이틀 제목

plt.subplot(223)
plt.plot(df3['x'],df3['y'],'*')
plt.title('ax3') # 서브 타이틀 제목

plt.subplot(224)
plt.plot(df4['x'],df4['y'],'d')
plt.title('ax4') # 서브 타이틀 제목

fig.suptitle('Amscombe', size=20) # 전체 타이틀
fig.tight_layout() # 간격을 알아서 맞춰줌
plt.show()
```
