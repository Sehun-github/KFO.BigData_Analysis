### 한글 폰트 사용 방법

```python
import matplotlib.pyplot as plt
```

```python
# 폰트 확인하기
import matplotlib.font_manager as fm
[f.name for f in fm.fontManager.ttflist if 'Malgun Gothic' in f.name]  
# 'Malgun Gothic' 부분에 Nanum으로 폰트 확인 가능
```

```python
# 한글폰트 지정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 한글폰트 사용 시 -기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False
```

### 색상
* color=색상으로 표현
* 색상이름 URL - https://matplotlib.org/2.0.2/examples/color/named_colors.html
* 자주 사용되는 색깔은 약자를 사용가능 (blue:b, green:g, red:r, cyan:c, magenta:m, yellow:y, black:k, white:w)
* hexcode로도 가능

```python
x = ['봄','여름','가을','겨울']
y = [19,30,15,-10]
plt.plot(x,y,color='r')
# plt.plot(x,y,color='#A566FF') # 헥스코드로도 가능
plt.show()
```

### 마커
* marker : 마커종류 (* . , o v ^ < > 1 2 3 4 s p * h H + x D d)
* markersize, ms : 마커사이즈
* markeredgecolor, mec : 마커 선 색깔
* markeredgewidth, mew : 마커 선 굵기
* markerfacecolor, mfc : 마커 내부 색깔

```python
x = ['봄','여름','가을','겨울']
y = [19,30,15,-10]
plt.plot(x,y, marker='D', ms='10', mec='b', mew='3', mfc='y')
plt.show()
```

### 선
* linestyle, ls : 선 스타일     
``'-'``    :      굵은선    
``'--'``    :     -- 선    
``'-.'``    :     -. 선    
``':'``     :     점선    

* linewidth, lw : 선 굵기

```python
x = ['봄','여름','가을','겨울']
y = [19,30,15,-10]
plt.plot(x,y, ls='-.', lw=3)
plt.show()
```

### 색상 & 마커 & 선스타일

```python
x = ['봄','여름','가을','겨울']
y = [19,30,15,-10]
plt.plot(x,y,'g*:', ms=10, mec='r', mfc='b', lw=3) # g*: 선색 green, 마크 *, 선스타일 점선 
plt.show()
```
