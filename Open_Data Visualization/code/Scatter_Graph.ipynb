## 스캐터
* 점의 크기
* 점의 색상
* 컬러맵 설정

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# 그래프에 한글 설정
mpl.rcParams['font.family'] = 'NanumSquare'

# 유니코드에서 그래프에 마이너스 기호 깨지는 문제 해결
mpl.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터
x = [1,5,6,9,10]
y = [1,5,3,9,7]
```

```python
plt.scatter(x,y)
plt.show()
```

### 점의 크기
* 전체 크기 설정 s = 숫자
* 개별 크기 설정 s = [크기,크기,크기]

```python
plt.scatter(x,y, s=[100,200,300,400,500])
plt.show()
```

### 점의 색상
* 전체 색상 설정 c = "색상"
* 개별 색상 설정 c = ["색상","색상","색상"]

```python
plt.scatter(x,y, s=[100,200,50,80,30], c = ['r','b','g','k','y'])
plt.show()
```

### 컬러맵 설정
* 컬러맵 지정 : c=y, cmap=컬러맵
* 컬러맵 : https://matplotlib.org/3.3.1/tutorials/colors/colormaps.html
* 컬러바 표시 : plt.colorbar()  

```python
# 컬러맵 종류
# plt.colormaps()
plt.scatter(x,y,c=y, cmap='magma')
plt.colorbar()
plt.show()
```

```python
# 지불금액과 팁의 관계
import seaborn as sns
tips = sns.load_dataset('tips')
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*50, alpha=0.5) # alpha: 투명도
plt.show()
```

```python
# 성별에 따른 color 컬럼 추가
def set_color(x):
    if x == 'Male':
        return 'blue'
    elif x == 'Female':
        return 'red'
    
tips['color'] = tips['sex'].apply(set_color)
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*50, alpha=0.5, c=tips['color'])
plt.show()
```
