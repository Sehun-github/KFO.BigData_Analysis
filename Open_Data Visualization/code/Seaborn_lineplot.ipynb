## Seaborn을 통한 선그래프 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
```

```python
# 데이터
flights = sns.load_dataset('flights')
flights.head(3)
```

## Lineplot
* `sns.lineplot(data=데이터프레임, x=x축컬럼, y=y축컬럼, estimator=통계함수)`<br>


```python
sns.lineplot(data=flights, x='year',y='passengers', ci=None, estimator='sum')
plt.show()
```

```python
# 연도-월별 승객수의 변화
sns.lineplot(data=flights, x='month', y='passengers', ci=None, hue='year')
plt.show()
```
