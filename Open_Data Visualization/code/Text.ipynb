## 텍스트 추가하기
* plt.text(x좌표, y좌표, 텍스트)
* rotation=회전각도
* ha : horizontal alignment &nbsp;  `['center'|'left'|'right']`
* va : vertical alignment  &nbsp;  `['center'|'top'|'bottom'|'baseline']`
* 텍스트 상자<br>
`bbox = {'boxstyle':상자스타일, 'fc':facecolor,'ec':edgecolor,...}`<br>
boxstyle : 'round'/'square'

```python
import pandas as pd
import matplotlib.pyplot as plt
# 그래프에 한글폰트 설정
plt.rcParams['font.family']='Malgun Gothic'

# 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot([1,2,3,4], 'ko')
plt.text(2.1,3,'(x:2,y:3)',ha='left', va='bottom', fontsize=12, rotation=45
        , bbox={'boxstyle': 'round', 'fc':'skyblue', 'ec':'blue', 'alpha':0.3})

plt.axhline(3, ls=':', alpha=0.5,lw=0.5)
plt.axvline(2, ls=':', alpha=0.5,lw=0.5)
plt.plot(2,3,'ro')
plt.show()
```

## 화살표와 텍스트 추가하기
* `plt.annotate('텍스트',xy=(화살표x,화살표y), xytext=(텍스트x,텍스트y), arrowprops=화살표속성(딕셔너리))`

* 화살표 속성<br>
`width`   
`headwidth`   
`headlength`    
`shrink` 

```python
plt.plot([1,2,3,4], 'ko')
plt.axhline(2, alpha=0.3,color='orange', lw=0.5,ls='--')
plt.axvline(1, alpha=0.3,color='orange',lw=0.5,ls='--')
plt.plot(1,2,'ro')

plt.annotate('(x:1,y:2)', xy=(1,2), xytext=(1.5,2.5), 
             arrowprops={'width':3,'headwidth':10,'headlength':10, 'shrink': 0.1,"fc":'r'},
            fontsize=12, color='r')
plt.show()
```
