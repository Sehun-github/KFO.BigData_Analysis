## 제목, 축 레이블, 그리드 설정
1. 제목
2. x축, y축 레이블
3. 그리드

```python
import matplotlib.pyplot as plt
# 그래프에 한글폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# 유니코드에서 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터 설정 및 기본 그래프 그리기
import pandas as pd
df = pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[80,78,75,73,70]})
plt.plot(df['월'],df['몸무게'], 'ro-', mec = 'b', mfc='b',lw=3, ms=7)
plt.show()
```

### 제목
* plt.title('제목')
* loc : 제목 위치('left','center','right')
* pad : 타이틀과 그래프와의 간격
* color : 폰트색상
* fontsize : 폰트사이즈
* fontweight : 폰트굵기('normal','bold','heavy','light','ultrabold','ultralight')
* fontfamily : 폰트

```python
import pandas as pd
df = pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[80,78,75,73,70]})
plt.plot(df['월'],df['몸무게'], 'ro-', mec = 'b', mfc='b',lw=3, ms=7)


plt.title("월별 몸무게 변화", loc='center',pad = 15, color = 'green', 
          fontsize= 20, fontweight='bold', fontfamily='NanumMyeongjo')


plt.show()
```

### x축, y축 레이블
* plt.xlabel('레이블')     
* plt.ylabel('레이블')
* loc : 위치(x축-'left','center','right' / y축-'bottom', 'center','top')
* labelpad : 레이블과 그래프와의 간격
* color : 폰트색상
* fontsize : 폰트사이즈
* fontfamily : 폰트

```python
import pandas as pd
df = pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[80,78,75,73,70]})
plt.plot(df['월'],df['몸무게'], 'ro-', mec = 'b', mfc='b',lw=3, ms=7)
plt.title("월별 몸무게 변화", loc='center',pad = 10, color = 'r', 
          fontsize= 20, fontweight='bold', fontfamily='NanumMyeongjo') # 제목


plt.xlabel('월', loc='right', labelpad = 10, color = 'darkorange',fontsize = 12, fontfamily = 'NanumMyeongjo') # xlabel
plt.ylabel('몸무게(kg)',loc='top', labelpad = 10, fontsize = 12, fontfamily = 'NanumMyeongjo') # ylabel


plt.show()
```

### 그리드
* plt.grid()
* axis : 그리드 방향
* ls-라인스타일, lw-굵기, color-색깔, alpha-투명도(0~1) 등의 속성을 지정할 수 있음

```python
import pandas as pd
df = pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[80,78,75,73,70]})
plt.plot(df['월'],df['몸무게'], 'ro-', mec = 'b', mfc='b',lw=3, ms=7)
plt.title("월별 몸무게 변화", loc='center',pad = 10, color = 'r', 
          fontsize= 20, fontweight='bold', fontfamily='NanumMyeongjo') # 제목
plt.xlabel('월', loc='right', labelpad = 10, color = 'darkorange',fontsize = 12, fontfamily = 'NanumMyeongjo') # xlabel
plt.ylabel('몸무게(kg)',loc='top', labelpad = 10, fontsize = 12, fontfamily = 'NanumMyeongjo') # ylabel


plt.grid(axis='y',color ='skyblue', alpha=0.5, ls=":") # 그리드


plt.show()
```
