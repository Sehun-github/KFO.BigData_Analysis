```python
df = sns.load_dataset("titanic")
df.head()
```

```python
# 다음 조건을 만족하는 코드를 입력하세요.
# 나이가 30살 이상 남자 승객 조건 필터링
# fare를 많이 낸 순서로 내림차순 정렬
# 상위 10개를 출력
```

```python
df[df['age'] >=30].sort_values('fare',ascending=False).iloc[:10]
```

```python
# 다음 조건을 만족하는 코드를 입력하세요.
# 나이가 20살 이상 40살 미만인 승객
# pclass가 1등급 혹은 2등급인 승객
# 열(column)은 survived, pclass, age, fare 만 나오게 출력

# 10개만 출력
```

```python
col = df['age'] >= 20 
col2 = df['age'] < 40 
col3 = df['pclass'] == 1 
col4 = df['pclass'] == 2 
```

```python
df.loc[(col & col2)&(col3|col4)].loc[:,['survived','pclass','age','fare']].iloc[:10]
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# tips 데이터셋 중 day가 금요일(Fri), 토요일(Sat) 만 필터링 합니다.
# tip이 $10보다 적게 낸 데이터만 필터링합니다.
# 컬럼은 total_bill, tip, smoker, time만 출력합니다.
# 상위 10개 행만 출력합니다.
```

```python
col = tips['day'].isin(['Fri','Sat'])
col2 = tips['total_bill'] < 10
```

```python
tips.loc[(col&col2)].loc[:,['total_bill', 'tip', 'smoker', 'time']]
```
