# 06. Groupby와 Pivot table(with타이타닉호 데이터)

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df['who'].value_counts()
```

```python
def transform_who(x):
    if x == 'man':
        return '남자'
    elif x == 'woman':
        return '여자'
    else:
        return '아이'
```

```python
df['who'].apply(transform_who)
```

```python
df['who'].apply(transform_who).value_counts()
```

```python
df['survived'] = df['survived'].apply(lambda x: '생존' if x == 1 else '사망')
```

```python
df.head()
```

## Groupby()

```python
df = sns.load_dataset('titanic')
df.groupby('sex').mean()
```

```python
df.groupby(['sex', 'pclass']).mean()
```

```python
df.groupby(['sex', 'pclass'])[['survived']].agg(['sum','count'])
```

```python
df.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean', 'sum'])
```

```python
df.groupby(['sex', 'pclass'])['survived'].mean().reset_index()
```

```python
df.pivot_table(index='who', values='survived')
```

```python
df = sns.load_dataset('titanic')
df.pivot_table(columns='who', values='survived')
```

```python
df.pivot_table(index=['who', 'pclass'], values='survived')
```

```python
df.pivot_table(index='who', columns='pclass', values='survived')
```

```python
df.pivot_table(index='who', columns='pclass', values='survived', aggfunc=['sum', 'mean']).reset_index
```
